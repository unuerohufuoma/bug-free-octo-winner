```python
df.drop(['date'], axis=1, inplace=True)
```

```python
# splitting data into x and y
x = df.drop(columns =['Appliances'])
y = df['Appliances']
```

```python
print(x)
```

```python
print(y)
```

```python
# normalise the dataset to a common scale using the min max scaler
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
normalised_df= pd.DataFrame(scaler.fit_transform(df),
columns=df.columns)
features_df = normalised_df.drop(columns=['Appliances'])
heating_target = normalised_df['Appliances']
```

```python
# split our data set into training and testing dataset. Recall that we had earlier segmented the features and target variable
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test =train_test_split(features_df, heating_target, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
linear_model=LinearRegression()
```

```python
# fit the model to the trainning data set
linear_model.fit(x_train, y_train)
```

```python
# obtain predictions
predicted_values = linear_model.predict(x_test)
```

```python
#MAE
from sklearn.metrics import mean_absolute_error
mae =mean_absolute_error(y_test, predicted_values)
round(mae, 3)
```

```python
from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(y_test, predicted_values))
round(rmse, 3)
```

```python
from sklearn.metrics import r2_score
r2_score = r2_score(y_test, predicted_values)
round(r2_score, 3)
```

```python
## Ridge Regression
# Import the ridge regression module from sklearn
from sklearn.linear_model import Ridge
```

```python
# Create ridge model
ridge = Ridge()
```

```python
# Train the model
ridge.fit(x_train, y_train)
```

```python
# Extract the model intercept value
b0 = float(ridge.intercept_)
```

```python
# Extract the model coefficient value
coeff = pd.DataFrame(ridge.coef_, x.columns, columns=['Coefficient'])
```

```python
print("Intercept:", float(b0))
```

```python
# Check out the coefficients
coeff
```
